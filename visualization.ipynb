```python
import torch

import scipy.io
import matplotlib.pyplot as plt

from pathlib import Path

plt.rcParams['figure.dpi'] = 200
plt.rcParams['savefig.dpi'] = 300
```

```python
# load input spike data
mat = scipy.io.loadmat(Path('data') / 'raw' / '01.mat') # animal 01
```

```python
# plot some sample images for each category
images_all = mat['images'].squeeze()
fig, ax = plt.subplots(1, 3, figsize=(8, 8))

fig_titles = ['Small image', 'Big image', 'Grating']
fig_idx = [538, 539, 600]
for i in range(3):
    ax[i].imshow(images_all[fig_idx[i]], cmap='gray')
    ax[i].set_xticks([])
    ax[i].set_yticks([])
    ax[i].set_title(fig_titles[i])

plt.savefig(Path('figures') / 'data_overview.png')
plt.show()
```
